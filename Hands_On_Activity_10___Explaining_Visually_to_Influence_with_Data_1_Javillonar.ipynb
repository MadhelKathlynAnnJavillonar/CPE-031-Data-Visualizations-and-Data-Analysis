**4. Procedures:**

Step 1. Import Libraries and Create Sample Data

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample sales dataset
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [200, 250, 180, 300, 280, 350],
    "Marketing_Spend": [50, 60, 45, 70, 65, 80]
}
df = pd.DataFrame(data)
df
```

Step 2. Know Your Audience — Compare Chart Styles

```python
# Executive audience: prefer clarity (simple bar)
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="steelblue")
plt.title("Executive View: Monthly Sales Overview", fontsize=13)
plt.show()

# Analyst audience: prefer detailed trends (line chart)
plt.figure(figsize=(7,4))
sns.lineplot(x="Month", y="Sales", data=df, marker="o", label="Sales")
sns.lineplot(x="Month", y="Marketing_Spend", data=df, marker="o", label="Marketing Spend")
plt.title("Analyst View: Sales vs Marketing Trend", fontsize=13)
plt.legend()
plt.show()
```

**Task 1:** Create one additional visualization suited for a different audience (e.g., a customer, investor, or teammate).

```python
df.set_index("Month")[["Sales", "Marketing_Spend"]].plot(kind="bar", stacked=True, figsize=(7, 4), color=["steelblue", "lightcoral"])

plt.title("Teammate View: Sales and Marketing Spend by Month", fontsize=13)
plt.xlabel("Month")
plt.ylabel("Amount ($)")
plt.xticks(rotation=45)
plt.show()
```

Step 3. Know Your Message — Highlight What Matters

```python
# Highlight the highest sales month (preattentive color cue)
highlight = df["Sales"].idxmax()

colors = ["gray" if i != highlight else "crimson" for i in range(len(df))]
plt.figure(figsize=(7,4))
plt.bar(df["Month"], df["Sales"], color=colors)
plt.title("Highlighting the Best Month in Sales", fontsize=13)
plt.show()
```

**Task 2:** Use color or annotations to emphasize a key insight (e.g., lowest point, sudden increase).

```python
highlight_lowest = df["Sales"].idxmin()

colors_lowest = ["gray" if i != highlight_lowest else "yellow" for i in range(len(df))]

plt.figure(figsize=(7,4))
bars = plt.bar(df["Month"], df["Sales"], color=colors_lowest)

plt.title("Highlighting the Lowest Sales Month", fontsize=13)
plt.xlabel("Month")
plt.ylabel("Sales ($)")

lowest_sales_value = df["Sales"].min()
plt.annotate(f"Lowest: {lowest_sales_value}",
             xy=(highlight_lowest, lowest_sales_value),
             xytext=(highlight_lowest, lowest_sales_value + 20),
             arrowprops=dict(arrowstyle="->", color="black"),
             fontsize=12, color="black", ha="center")

plt.show()
```

Step 4. Storytelling with Charts — Beginning, Middle, End

```python
# Beginning: Context
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="skyblue")
plt.title("Beginning: Sales Started Steady", fontsize=13)
plt.show()

# Middle: Change or Challenge
plt.figure(figsize=(7,4))
sns.lineplot(x="Month", y="Sales", data=df, color="orange", marker="o")
plt.title("Middle: A Drop in March Due to Low Marketing", fontsize=13)
plt.show()

# End: Resolution
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="limegreen")
plt.title("End: Recovery and Growth After March", fontsize=13)
plt.show()
```

**Task 3:** Write one or two lines describing your data story from beginning → middle → end.

Sales began consistently strong at the start of the year, reflecting a steady performance. In March, a decline occurred due to reduced marketing efforts, but with increased marketing spend afterward, sales rebounded and showed clear growth by June.


Step 5. Bringing It All Together — Presentation Design

```python
# Simple combined story chart
plt.figure(figsize=(8,5))
sns.lineplot(x="Month", y="Sales", data=df, marker="o", linewidth=2, color="royalblue", label="Sales")
sns.lineplot(x="Month", y="Marketing_Spend", data=df, marker="o", linestyle="--", color="orange", label="Marketing Spend")

plt.fill_between(df["Month"], df["Sales"], alpha=0.1, color="royalblue")
plt.title("From Effort to Impact: Marketing Spend Drives Sales Growth", fontsize=13)
plt.legend()
plt.show()
```

**5. Supplementary Activity:**

1. Use a small dataset available online (kaggle, google dataset, etc.).

2. Visualize it using a bar chart.

3. Highlight one important data point (e.g., the highest or lowest value) using a distinct color.

4. Add a title and annotation that tell the story behind the data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/drive/MyDrive/CPE 031 JAVILLONAR/NationalNames.csv')

latest_year = df['Year'].max()
df_recent = df[df['Year'] == latest_year].sort_values('Count', ascending=False).head(10).copy()

highlight_idx = df_recent['Count'].idxmax()
colors = ['lightgray' if i != highlight_idx else 'crimson' for i in df_recent.index]

plt.figure(figsize=(10,6))
barplot = sns.barplot(x='Count', y='Name', data=df_recent, palette=colors)

plt.title(f'Top 10 Baby Names in {latest_year} — Highlighting the Most Popular', fontsize=16, weight='bold')
plt.xlabel('Number of Babies', fontsize=12)
plt.ylabel('Name', fontsize=12)

for i, (count, name) in enumerate(zip(df_recent['Count'], df_recent['Name'])):
    barplot.text(count + 50, i, str(count), va='center', fontsize=10, color='black', weight='bold')

top_name = df_recent.loc[highlight_idx, 'Name']
top_count = df_recent.loc[highlight_idx, 'Count']

plt.annotate(
    f"Most popular: {top_name} ({top_count} babies)",
    xy=(top_count, df_recent.index.get_loc(highlight_idx)),
    xytext=(top_count*0.6, df_recent.index.get_loc(highlight_idx)+0.5),
    arrowprops=dict(arrowstyle='->', color='black', lw=2),
    fontsize=12, color='black', weight='bold'
)

plt.tight_layout(pad=3)

plt.show()
```
